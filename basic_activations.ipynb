```python
import torch
import matplotlib.pyplot as plt
from functools import partial
from ipywidgets import interact
from math import e
```

```python
def relu(m, b, x):
  y = x * m + b
  return torch.clip(y, 0.)
```

```python
def elu(z,alpha):
  return torch.where(z >= 0, z, alpha * (torch.exp(z) - 1))
```

```python
def f(m1, b1, m2, b2, a1, x):
  """
    Sums all the functions, basically mimcs what an ML model or torch does with neurons activations.
  """
  return relu(m1, b1, x) + relu(m2, b2, x) + elu(x, a1)
```

```python
@interact(m1=1.5, b1=1.5, m2=1.5, b2=1.5, a1=1.5)
def plot_f(m1, b1, m2, b2, a1):
  x = torch.linspace(-10, 10, 100)
  y = f(m1, b1, m2, b2, a1, x)
  plt.figure(figsize=(10, 6))
  plt.plot(x, y, 'b-', label='f(x) = x^2 + 2x + 1')
  plt.title('Plot of f(x) = x^2 + 2x + 1')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.grid(True)
  plt.legend()
  plt.axhline(y=0, color='k')
  plt.axvline(x=0, color='k')
  plt.show()
```
